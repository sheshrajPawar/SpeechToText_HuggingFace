```python
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
import numpy as np
```

```python
from IPython.display import display
```

```python
tokenizer =  Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
from IPython.display import Audio

# Create an audio object and play it
audio = Audio("audio1.wav", autoplay=True)
display(audio)
```

```python
import librosa
from transformers import Wav2Vec2Tokenizer

# Load the tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")

# Load the audio file (make sure the path is correct)
audio_path = "audio1.wav"
audio, sample_rate = librosa.load(audio_path, sr=16000)  # Load with the correct sample rate

# Tokenize the audio input
input_values = tokenizer(audio, return_tensors='pt').input_values

# Display the tokenized input values
print(input_values)
```

```python
logits = model(input_values).logits
logits
```

```python
predicted_ids = torch.argmax(logits, dim= -1)
```

```python
transcriptions = tokenizer.decode(predicted_ids[0]) 
```

```python
transcriptions
```
